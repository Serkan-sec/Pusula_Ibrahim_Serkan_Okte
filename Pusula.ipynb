```python
def custom_split(text):

    parts = []
    inside_paren = False
    current_part = ""

    for char in text:
        if char == '(':
            inside_paren = True
            current_part += char
        elif char == ')':
            inside_paren = False
            current_part += char
        elif char == ',' and not inside_paren:
            parts.append(current_part.strip())
            current_part = ""
        else:
            current_part += char


    if current_part:
        parts.append(current_part.strip())

    return parts

df_clean['Ilac_Adi'] = df_clean['Ilac_Adi'].apply(custom_split)
```
